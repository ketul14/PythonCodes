```python
# question
l1 = [1,4,5,7]
l2 = [2,3,7,9,12,14,17]

# method 1 
def merge1(list1, list2):
    
    n1 = len(list1)
    n2 = len(list2)
    
    if n1 >= n2:
        output = list1
        for i in range(n2):
            output.append(list2[i])
    else:
        output = list2
        for i in range(n1):
            output.append(list1[i])
    
    output.sort()
    
    return output
        
    
# method 2
def merge2(list1, list2):
    return sorted(list1 + list2)


# method 3
def merge3(list1, list2):
    
    n1 = len(list1)
    n2 = len(list2)
    output = []
    i,j = 0,0
    while i < n1 and j < n2:
        if list1[i] <= list2[j]:
            output.append(list1[i])
            i+= 1
        else:
            output.append(list2[j])
            j+= 1
    return output + list1[i:] + list2[j:]

print(f' method 1: ', merge1(l1, l2))
print(f' method 2: ', merge2(l1, l2))
print(f' method 3: ', merge3(l1, l2))
    
```
