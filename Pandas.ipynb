```python
####### Data Cleansing
# rename columns 
# rename all columns
df.columns = ['Category_id', 'Date', 'ApplicationCount', 'PendingCount',
       'ApprovedCount', 'FundedCount', 'WithdrawnCount', 'TurndownCount',
       'LoanCount', 'Inventory', 'Avg_LoanAmount']
df.columns
```

```python
# rename individual column
# 1. By dictionary
df.rename(columns = {'Category_id': 'Category_id1'}, inplace = True)

# 2. By Index string replace (Advantage is Regex)
df.columns = df.columns.str.replace('Category_id1', 'Category_id2')

# 3. List comprehensive 
df.columns = ['Category_ID' if x == 'Category_id2' else x for x in df.columns]
df.columns
```

```python
# Drop NAs
# Drop all rows with NAs
df1 = df.copy()
df1.dropna(axis=0) # Axis = 0 for rows, axis =1 for Columns
```

```python
# Drop NAs with thresold
df1.dropna(axis=0,thresh=3)
```

```python
# replace NAs with value
df1.fillna('Other')
```

```python
# replace NAs with mean
df.fillna(df.ApplicationCount.mean())
```

```python
# set index and reset index
df.set_index('Category_ID', inplace = True)
df
```

```python
# reset index
df.reset_index(inplace = True)
df
```

```python
# Filtering data
df[df['ApplicationCount'] > 100]
```

```python
# Multiple condition filtering
df[ (df['ApplicationCount'] > 100 )  & ( df['ApplicationCount'] < 200) ]
```
