```python
####### Data Cleansing
# rename columns 
# rename all columns
df.columns = ['State/UTs', 'Total Cases', 'Active', 'Discharged', 'Deaths',
       'Active Ratio (%)', 'Discharge Ratio (%)', 'Death Ratio (%)']
df.columns
```

```python
# rename individual column
# 1. By dictionary
df.rename(columns = {'State/UTs': 'State'}, inplace = True)

# 2. By Index string replace (Advantage is Regex)
df.columns = df.columns.str.replace('State', 'State')

# 3. List comprehensive 
df.columns = ['States' if x == 'State' else x for x in df.columns]
df.columns
```

```python
# Check null values

# check all columns
df.isnull()
```

```python
# check single columns
df[['States']].isnull()
```

```python
# check not null
df[['States']].notnull()
```

```python
df.at[5, 'States'] == None
```

```python
# Drop NAs
# Drop all rows with NAs
df1 = df.copy()
df1.dropna(axis=0) # Axis = 0 for rows, axis =1 for Columns
```

```python
# Drop NAs with thresold
df1.dropna(axis=0,thresh=3)
```

```python
# replace NAs with value
df1.fillna('Other')
```

```python
# replace NAs with mean
df.fillna(df['Active'].mean())
```

```python
# set index and reset index
df.set_index('States', inplace = True)
df
```

```python
# reset index
df.reset_index(inplace = True)
df
```

```python
# Filtering data
df[df['Active'] > 5000]
```

```python
# Multiple condition filtering
df[ (df['Active'] > 5000 )  & ( df['Active'] < 10000) ]
```
